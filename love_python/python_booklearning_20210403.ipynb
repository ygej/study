### 테스트 데이터 준비

````python
# Import library
import pandas as pd
import numpy as np

# Load data
titanic_df = pd.read_csv('D:/data/titanic/train.csv')
print(titanic_df.shape)
titanic_df.head()
````

### 1. 데이터 셀렉션 및 필터링

판다스 DataFrame은 크게 세 가지 요소로 구성되어 있습니다. 
1. 데이터의 행(row)을 가져오는 기준인 인덱스(index)
2. 데이터의 열을 가져오는 기준인 컬럼(column)
3. 데이터의 값을 가져오는 기준인 값(values)

각각을 ```.index```, ```.columns```, 그리고 ```.values```를 활용해 불러올 수 있습니다.

#### 1.1 행(row) 가져오기
1) DataFrame에서 ```대괄호[]```를 사용합니다.

2) ```.loc```(locate의 약자)라는 문법을 사용합니다.

3) 대괄호 안에 ```인덱스(index)```를 넣습니다. 정확히는 인덱스의 명칭을 넣어줍니다!

````python
# 데이터 타입은 Series 입니다.
print(type(titanic_df.loc[109]))

# titanic_df의 109 인덱스 가져오기
titanic_df.loc[109]
````

````python
# 여려 개의 행(row) 접근하기
# 대괄호를 '[[]]' 두 번 사용합니다.
# 안쪽 대괄호를 파이썬의 리스트(list)라고 간주합니다.

# 데이터 타입은 DataFrame 입니다.
print(type(titanic_df.loc[[109,720]]))

# titanic_df의 109, 720 인덱스 가져오기
titanic_df.loc[[109,720]]
````

````python
# 안쪽 대괄호의 인덱스를 리스트로 지정해도 결과는 같습니다.
my_list = [109,720]

titanic_df.loc[my_list]
````

#### 1.2 열(column) 가져오기

1) DataFrame에서 ```대괄호[]```를 사용합니다.

2) 이 대괄호에 ```문자열```을 집어넣으면 문자열에 해당하는 열(column)을 반환합니다.


````python
# 데이터 타입은 Series 입니다.
print(type(titanic_df['Name']))

# titanic_df의 'Name' 열 가져오기
titanic_df['Name']
````

````python
# 여려 개의 열(column) 접근하기
# 대괄호를 '[[]]' 두 번 사용합니다.
# 안쪽 대괄호를 파이썬의 리스트(list)라고 간주합니다.

# 데이터 타입은 DataFrame 입니다.
print(type(titanic_df[['Name', 'Age']]))

# titanic_df의 'Name', 'Age' 열 가져오기
titanic_df[['Name', 'Age']]
````

#### 1.3 행(row), 열(column) 동시에 가져오기

1) ```.loc[<<index>>][<column>>]``` 으로 행렬을 동시에 가져올 수 있지만 추천하지 않습니다..

2)  ```.loc[<<index>>, <column>>]``` 을 추천합니다


````python
# titanic_df의 인덱스 109의 'Name' 가져오기
titanic_df.loc[109, 'Name']
````

````python
# 여러개의 행(row)의 열(column) 가져오기
titanic_df.loc[[109,720], 'Name']
````

````python
# 여러개의 여러개의 열(column)의 행(row) 가져오기
titanic_df.loc[109, ['Name', 'Age']]
````

````python
# 여러개의 행(row)의 여러개의 열(column) 가져오기
titanic_df.loc[[109,720], ['Name','Age']]
````

#### 1.4 행(row), 열(column) 순서대로 가져오기

1) ```.iloc``` 문법을 사용합니다.

2)  ```대괄호[]``` 안에 가져올 데이터의 순서를 넣습니다.

````python
# 데이터 타입은 Series 입니다.
print(type(titanic_df.iloc[0]))

# titanic_df의 첫번째(0번째) 행(row) 가져오기
titanic_df.iloc[0]
````

````python
# titanic_df의 첫번째(0) 행(row)의 세번째(2) 열(column) 가져오기
titanic_df.iloc[0,2]
````

````python
# titanic_df의 첫번째(0), 다섯번째(4) 행(row)의 세번째(2), 여섯번째(5) 열(column) 가져오기
titanic_df.iloc[[0,4],[2,5]]
````

#### 1.5 색인(indexing)

1) ```비교연산자(=, >, < 등)``` 문법을 사용합니다.

````python
# titanic_df에 특정 값(Moran, Miss. Bertha)이 존재하는지 여부를 확인할 수 있습니다.
titanic_df[titanic_df["Name"] == "Moran, Miss. Bertha"]
````

````python
# titanic_df에 특정 값(Moran, Miss. Bertha)외의 값을 확인할 수 있습니다.
titanic_df[titanic_df["Name"] != "Moran, Miss. Bertha"]
````

````python
# 정수형, 실수형은은 크고 작음으로 데이터를 색인할 수 있습니다.
titanic_df[titanic_df["Age"] > 70]
````

````python
# ``&(and) 기호와 |(or) 기호```를 활용하여 특정 조건을 지정하여 색인할 수 있습니다.
titanic_df[(titanic_df["Age"] > 70) & (titanic_df["Pclass"] == 1) ]
````

````python
# 조건이 복잡해지면 이렇게 여러 개의 조건으로 나눠서 사용할 수 있습니다.
old = titanic_df["Age"] > 70
pclass_1 = titanic_df["Pclass"] == 1

titanic_df[old & pclass_1]
````

### 2. 정렬,Aggregation함수, Groupby 적용

#### 2.1 정렬하기

1) index로 정렬합니다. ```.sort_index```

2) column으로 정렬합니다. ```.sort_values```


````python
# index를 기준으로 정렬하기
# 숫자는 작은 순(0 ~ 9)으로 알파벳은 A부터 Z까지 정렬합니다.
# 대소문자가 섞여있을 경우 대문자가 앞으로, 소문자가 뒤로 옵니다. 
titanic_df.sort_index()
````

````python
# 정렬 방식을 오름차순(ascending)이 아닌 내림차순(descending)으로 할 경우, ascending 옵션에 False를 주면 됩니다.
titanic_df.sort_index(ascending=False)
````

````python
# 만일 행(row)이 아닌 열(column)을 정렬하고 싶다면 axis = 1 옵션을 줍니다.
# 이렇게 하면 컬럼(column) 값을 낮은 순으로(숫자면 0 ~ 9, 알파벳은 A ~ Z) 정렬합니다.
titanic_df.sort_index(axis=1)
````

````python
# 컬럼값을 기준으로 정렬하기
# by 옵션에는 컬럼명이 들어갑니다.
# 이럴 경우 해당 컬럼값이 작은 순(숫자는 0 ~ 9, 알파벳은 A ~Z)으로 정렬합니다.
titanic_df.sort_values(by="Name")
````

````python
# 정렬 방식을 오름차순(ascending)이 아닌 내림차순(descending)으로 할 경우, ascending 옵션에 False를 주면 됩니다.
titanic_df.sort_values(by="Name", ascending=False)
````

````python
# 여러 개의 컬럼값을 기준으로 정렬하고 싶다면 by 옵션에 여러 개의 컬럼명을 리스트로 만들어 넣으면 됩니다.
# 1) Name 컬럼값이 낮은 순으로 정렬한 뒤, 2) Age 컬럼값이 낮은 순으로 정렬합니다.
titanic_df.sort_values(by = ["Name", "Age"])
````
